```python
import numpy as np

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
axes[0,0].plot(x1,y1)
axes[0, 0].set_title('Line 1')
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line 2')

# Plot 3 : Line 3
axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line 3')

# Plot 4 : Line 4
axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line 4')
```

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
plt.bar(company,profit)
plt.xlabel('Company Name')
plt.ylabel('Profit')
plt.title('Profit Comparison for various Companies')
plt.show()
```

```python
import numpy as np

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
# Importing Matplotlib
import matplotlib.pyplot as plt

# Combine the data into a list
data = [box1, box2]

# Create the box plot
fig, ax = plt.subplots()
ax.boxplot(data,patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot of Two Datasets')

# Display the plot
plt.show()
```
